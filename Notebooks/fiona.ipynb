```python
cause_mapping = {
    'UNABLE TO DETERMINE': 'Unknown',

    # Driver Error
    'FAILED TO REDUCE SPEED TO AVOID CRASH': 'Driver Error',
    'IMPROPER OVERTAKING/PASSING': 'Driver Error',
    'FOLLOWING TOO CLOSELY': 'Driver Error',
    'FAILING TO YIELD RIGHT-OF-WAY': 'Driver Error',
    'DISREGARDING TRAFFIC SIGNALS': 'Driver Error',
    'DISREGARDING STOP SIGN': 'Driver Error',
    'IMPROPER LANE USAGE': 'Driver Error',
    'IMPROPER TURNING/NO SIGNAL': 'Driver Error',
    'DRIVING SKILLS/KNOWLEDGE/EXPERIENCE': 'Driver Error',
    'DISREGARDING OTHER TRAFFIC SIGNS': 'Driver Error',
    'IMPROPER BACKING': 'Driver Error',
    'DISTRACTION - FROM INSIDE VEHICLE': 'Driver Error',

    # External Factors
    'WEATHER': 'External Factors',
    'ROAD ENGINEERING/SURFACE/MARKING DEFECTS': 'External Factors',
    'VISION OBSCURED (SIGNS, TREE LIMBS, BUILDINGS, ETC.)': 'External Factors',
    'DISTRACTION - FROM OUTSIDE VEHICLE': 'External Factors',

    # Vehicle Defect
    'EQUIPMENT - VEHICLE CONDITION': 'Vehicle Defect',

    # Other
    'OTHER': 'Other'
}

df_clean['Crash_Cause'] = df_clean['prim_contributory_cause']\
                            .map(cause_mapping)\
                            .fillna('Other')

# Drop original 
df_clean = df_clean.drop(columns=['prim_contributory_cause'])

df_clean['Crash_Cause'].value_counts()
```

```python
X = df_clean.drop(columns=['Crash_Cause','prim_contributory_cause'], errors = 'ignore')
y = df_clean['Crash_Cause']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.25,
    random_state=42,
    stratify=y
)

X_train.shape, X_test.shape
```

```python
categorical_features = X_train.select_dtypes(include=['object']).columns
numeric_features = X_train.select_dtypes(exclude=['object']).columns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
# from xgboost import XGBClassifier

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    ConfusionMatrixDisplay,
    accuracy_score,
    f1_score
)

from sklearn.inspection import permutation_importance
```

```python
# Group rare categories → much fewer one-hot columns
rare_threshold = 500   # adjust if needed (higher = fewer columns)

for col in ['street_name', 'first_crash_type', 'trafficway_type', 'prim_contributory_cause_group']:  # add your high-cardinality columns
    if col in X_train.columns:
        counts = X_train[col].value_counts()
        common = counts[counts >= rare_threshold].index
        X_train[col] = X_train[col].where(X_train[col].isin(common), 'Rare/Other')
        X_test[col]  = X_test[col].where(X_test[col].isin(common), 'Rare/Other')
        print(f"Reduced {col}: {len(common)} common categories kept")
```

```python
print("\nUpdated categorical uniques after reductions/drops:")
print(X_train.select_dtypes(include="object").nunique().sort_values(ascending=False).head(10))
```

```python
high_card = ['location', 'street_name', 'date_police_notified']
cat_reduced = [c for c in categorical_features if c not in high_card]
```

```python


categorical_features = X_train.select_dtypes(include=['object']).columns
numeric_features = X_train.select_dtypes(exclude=['object']).columns

numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")), # handles missing values
    ("ohe", OneHotEncoder(handle_unknown="ignore", sparse_output=False)) # encoding categories
])

preprocessor = ColumnTransformer(transformers=[
    ("num", numeric_transformer, numeric_features),
    ("cat", categorical_transformer, cat_reduced)
])
```

```python
def evaluate(name, pipe):
    print(f"\n{'='*80}")
    print(f"Training {name}...")
    
    pipe.fit(X_train, y_train)
    preds = pipe.predict(X_test)

    acc = accuracy_score(y_test, preds)
    macro_f1 = f1_score(y_test, preds, average='macro')

    print(name)
    print(f"Accuracy: {acc:.4f}")
    print(f"Macro F1: {macro_f1:.4f}")
    print("\nClassification Report:")
    print(classification_report(y_test, preds))

    cm = confusion_matrix(y_test, preds, labels=pipe.named_steps['model'].classes_)
    ConfusionMatrixDisplay(cm, display_labels=pipe.named_steps['model'].classes_).plot(xticks_rotation=45)
    plt.title(f"Confusion Matrix - {name}")
    plt.show()

    return {"Model": name, "Accuracy": acc, "Macro_F1": macro_f1}
```

```python
results = []
```

```python
# Model Pipeline
tree_pipe = Pipeline(steps=[
    ("preprocess", preprocessor),
    ("model", DecisionTreeClassifier(max_depth=5, min_samples_leaf=100, random_state=42,class_weight='balanced'))
])

dt_param_grid = {
    'model__max_depth': [5, 10, 15, None],
    'model__min_samples_split': [2, 5, 10],
    'model__min_samples_leaf': [20, 50, 100],
    'model__max_features': [None, 'sqrt', 'log2']
}

results.append(evaluate("Decision Tree", tree_pipe))
```

```python


logreg_pipe = Pipeline(steps=[
    ("preprocess", preprocessor),
    ("model", LogisticRegression(
        max_iter=1000,
        class_weight='balanced',
        random_state=42,
        n_jobs=-1
         ))
])

lr_param_grid = {
    'model__C': [0.01, 0.1, 1.0, 10.0],           
    'model__solver': ['lbfgs', 'liblinear'],      
    'model__penalty': ['l2']                       
    
}

results.append(evaluate("Logistic Regression", logreg_pipe))
```

```python
# Random Forest
rf_pipe = Pipeline([
    ('preprocess', preprocessor),
    ('model', RandomForestClassifier(n_estimators=50, max_depth=12, min_samples_leaf=80, class_weight='balanced_subsample', random_state=42, n_jobs=-1))
])

rf_param_grid = {
    'model__n_estimators': [50, 100, 150],
    'model__max_depth': [10, 15, None],
    'model__min_samples_split': [2, 5],
    'model__min_samples_leaf': [50, 80, 100],
    'model__max_features': ['sqrt', 'log2']
}


results.append(evaluate("Random Forest", rf_pipe))
```

```python
# Common settings
cv = 3
scoring = 'f1_weighted'   
n_jobs = -1

print("Starting hyperparameter tuning for your three models...\n")
```

```python
# Tuned Decision Tree
print("Tuning Decision Tree...")
dt_search = GridSearchCV(
    tree_pipe,
    param_grid=dt_param_grid,
    cv=cv,
    scoring=scoring,
    n_jobs=n_jobs,
    verbose=1
)
dt_search.fit(X_train, y_train)

print("Best DT params:", dt_search.best_params_)
print(f"Best DT CV score ({scoring}): {dt_search.best_score_:.4f}\n")

results.append(evaluate("Decision Tree (Tuned)", dt_search.best_estimator_))
```

```python

print("Tuning Logistic Regression...")
lr_search = GridSearchCV(
    logreg_pipe,
    param_grid=lr_param_grid,
    cv=cv,
    scoring=scoring,
    n_jobs=n_jobs,
    verbose=1
)
lr_search.fit(X_train, y_train)

print("Best LR params:", lr_search.best_params_)
print(f"Best LR CV score ({scoring}): {lr_search.best_score_:.4f}\n")

results.append(evaluate("Logistic Regression (Tuned)", lr_search.best_estimator_))

```

```python

print("Tuning Random Forest...")
rf_search = GridSearchCV(
    rf_pipe,
    param_grid=rf_param_grid,
    cv=cv,
    scoring=scoring,
    n_jobs=n_jobs,
    verbose=1
)
rf_search.fit(X_train, y_train)

print("Best RF params:", rf_search.best_params_)
print(f"Best RF CV score ({scoring}): {rf_search.best_score_:.4f}\n")

results.append(evaluate("Random Forest (Tuned)", rf_search.best_estimator_))
```
